## Comprehensive Normal Traffic Analysis

Let's identify all legitimate/normal applications that should be used for training the anomaly detector.


```python
# Comprehensive list of normal/legitimate applications
# These are trusted applications that should be considered "normal" traffic

# Current normal apps (limited)
current_normal_apps = ['GOOGLE', 'YOUTUBE', 'FACEBOOK', 'MICROSOFT', 'CLOUDFLARE', 'AMAZON']

# Expanded list of legitimate applications
# Categories:
# - Major tech companies/services
# - Cloud services
# - Communication platforms
# - Content delivery
# - Operating system updates
# - Email services
# - Social media
# - Productivity tools

expanded_normal_apps = [
    # Current apps
    'GOOGLE', 'YOUTUBE', 'FACEBOOK', 'MICROSOFT', 'CLOUDFLARE', 'AMAZON',
    
    # Email services
    'GMAIL', 'YAHOO', 'MSN',
    
    # Cloud storage & services
    'DROPBOX', 'OFFICE_365',
    
    # Communication
    'SKYPE',
    
    # Social media
    'TWITTER',
    
    # Operating system updates
    'WINDOWS_UPDATE', 'APPLE',
    
    # Content delivery
    'CONTENT_FLASH',
    
    # Standard protocols (if they represent legitimate traffic)
    'HTTP', 'HTTPS', 'SSL',
]

# Check which apps exist in the dataset
if 'ProtocolName' in df.columns:
    all_protocols = df['ProtocolName'].unique()
    
    print("=" * 80)
    print("NORMAL TRAFFIC ANALYSIS")
    print("=" * 80)
    
    print(f"\nCurrent normal apps (6 apps):")
    current_counts = {}
    for app in current_normal_apps:
        if app in all_protocols:
            count = len(df[df['ProtocolName'] == app])
            current_counts[app] = count
            print(f"  {app:20s}: {count:>10,} samples")
        else:
            print(f"  {app:20s}: NOT FOUND")
    
    print(f"\nTotal samples in current normal apps: {sum(current_counts.values()):,}")
    print(f"Percentage of total dataset: {sum(current_counts.values())/len(df)*100:.2f}%")
    
    print(f"\n" + "=" * 80)
    print("EXPANDED NORMAL APPS ANALYSIS")
    print("=" * 80)
    
    expanded_counts = {}
    for app in expanded_normal_apps:
        if app in all_protocols:
            count = len(df[df['ProtocolName'] == app])
            expanded_counts[app] = count
            print(f"  {app:20s}: {count:>10,} samples")
        else:
            print(f"  {app:20s}: NOT FOUND")
    
    print(f"\nTotal samples in expanded normal apps: {sum(expanded_counts.values()):,}")
    print(f"Percentage of total dataset: {sum(expanded_counts.values())/len(df)*100:.2f}%")
    
    print(f"\n" + "=" * 80)
    print("ALL PROTOCOLS IN DATASET")
    print("=" * 80)
    print("\nProtocols sorted by count (top 30):")
    protocol_counts = df['ProtocolName'].value_counts()
    for i, (protocol, count) in enumerate(protocol_counts.head(30).items(), 1):
        is_normal = protocol in expanded_normal_apps
        marker = "✓" if is_normal else " "
        print(f"{i:2d}. {marker} {protocol:25s}: {count:>10,} samples")
    
    print(f"\n" + "=" * 80)
    print("RECOMMENDED NORMAL APPS LIST")
    print("=" * 80)
    print("\nBased on the analysis, here are the recommended normal apps:")
    recommended_normal = [app for app in expanded_normal_apps if app in all_protocols]
    print(f"\nRecommended normal apps ({len(recommended_normal)} apps):")
    for app in recommended_normal:
        print(f"  '{app}',")
    
    print(f"\nTotal samples: {sum(expanded_counts.values()):,}")
    print(f"Percentage: {sum(expanded_counts.values())/len(df)*100:.2f}%")
    
else:
    print("ProtocolName column not found in dataset")
```

## Correlation Analysis (Sample Features)


```python
# Select numeric columns for correlation
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numeric_cols) > 0:
    # Sample a subset for visualization
    sample_cols = numeric_cols[:20] if len(numeric_cols) > 20 else numeric_cols
    corr_matrix = df[sample_cols].corr()
    
    plt.figure(figsize=(14, 12))
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0, square=True, linewidths=0.5)
    plt.title('Feature Correlation Matrix (Sample)')
    plt.tight_layout()
    plt.show()
```

```python
# Show distribution using ProtocolName instead of numeric codes
if 'ProtocolName' in df.columns:
    plt.figure(figsize=(12, 18))
    protocol_counts = df['ProtocolName'].value_counts()
    sns.countplot(y=df['ProtocolName'], order=protocol_counts.index)
    plt.title('Distribution of Applications (ProtocolName)')
    plt.xlabel('Count')
    plt.ylabel('Application')
    plt.tight_layout()
    plt.show()
    
    print(f"\nTotal unique protocols: {df['ProtocolName'].nunique()}")
    print(f"\nAll ProtocolName values and their counts:")
    print(protocol_counts)
else:
    print("ProtocolName column not found. Using L7Protocol instead.")
    plt.figure(figsize=(12, 18))
    sns.countplot(y=df['L7Protocol'])
    plt.title('Distribution of Applications (L7Protocol - numeric codes)')
    plt.xlabel('Count')
    plt.ylabel('Application Code')
    plt.tight_layout()
    plt.show()
```
